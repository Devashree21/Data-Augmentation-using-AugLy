<p style="font-family: Arials; font-size: 20px;text-align: center;; font-style: normal;line-height:1.3">AugLy is a recently launched Python open-source project by Facebook for data augmentation. The goal is to increase the robustness of AI models during training and evaluation. The library includes four modes (audio, video, image, and text) and provides over 100 data augmentation methods.</p>

# [2. Importing libraries and Image](#2)

```python
import os
import augly.image as imaugs
import augly.utils as utils
from IPython.display import display
```

### <h3 style="background-color:#48cae4;color:black;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>Image Augmentation</centre></strong></h3>

```python
# Image path setting 
input_img = '../input/image-bird/bird.jpg'
```

# [3. Image Scaling ](#3)

```python
# Image Scaling with small factor value
input_img0 = imaugs.scale(input_img, factor=0.1)
display(input_img0)
```

```python
# Image Scaling with large factor value
input_img1 = imaugs.scale(input_img, factor=0.3)
display(input_img1)
```

# [4. Image Blurring ](#4)

```python
# Image Blurring
input_img2 = imaugs.blur(input_img1, radius = 4.0)
display(input_img2)
```

# [5. Image Brightness & Darkness ](#5)

```python
# Increasing Image Brightness
input_img3 = imaugs.brightness(input_img1,factor=1.5)
display(input_img3)
```

```python
# Making Image Darker
input_img4 = imaugs.brightness(input_img1,factor=0.3)
display(input_img4)
```

# [6. Image Contrast ](#6)

```python
# Changing Image Contrast
input_img5 = imaugs.contrast(input_img1,factor=2.7)
display(input_img5)
```

# [7. Image Flipping - Vertical and Horizontal ](#7)

```python
#hflip
input_img6 = imaugs.hflip(input_img1)
display(input_img6)
```

```python
#vflip
input_img7 = imaugs.vflip(input_img1)
display(input_img7)
```

pip install ipyplot

```python
import warnings
warnings.filterwarnings("ignore")

import ipyplot
images_list=[input_img1, input_img6, input_img7]

labels = ['Original', 'Horizontally flipped', 'Vertically flipped']

ipyplot.plot_images(images_list,labels=labels,img_width=150)
```

# [8. Image Grayscale ](#8)

```python
#Grayscale image
input_img8 = imaugs.grayscale(input_img1)
display(input_img8)
```

# [9. Degrading Image Pixels ](#9)

```python
# Degrading Image Pixels
input_img9 = imaugs.shuffle_pixels(input_img1, factor=0.3)
display(input_img9)
```

# [10. Image Skewing ](#10)

```python
#skew
input_img10 = imaugs.skew(input_img1)
display(input_img10)
```

# [11. Image Perspective Transform ](#11)

```python
# Image Perspective Transform
aug= imaugs.PerspectiveTransform(sigma=20.0)
input_img11=aug(input_img1)
display(input_img11)
```

# [12. Changing Aspect Ratio ](#12)

```python
#Changing Aspect Ratio
aug = imaugs.RandomAspectRatio()
input_img12=aug(input_img1)
display(input_img12)
```

# [13. Image Cropping ](#13)

```python
# Cropping Image
input_img13 = imaugs.crop(input_img1,x1=0.2,x2=0.8,y1=0.2,y2=0.8)
display(input_img13)
```

# [14. Creating meme ](#14)

```python
# Creating a meme
display(
    imaugs.meme_format(
        input_img1,
        caption_height=75,
        meme_bg_color=(0, 0, 0),
        text_color=(255, 255, 255),
    )
)
```

# [15. Applying Transformations ](#15)

```python
# Applying several transformations together to create a new image
aug = imaugs.Compose(
    [
        imaugs.Saturation(factor=0.7),
        imaugs.OverlayOntoScreenshot(
            template_filepath=os.path.join(
                utils.SCREENSHOT_TEMPLATES_DIR, "web.png"
            ),
        ),
        imaugs.Scale(factor=0.9),
    ]
)
display(aug(input_img1))
```

# [19. Text Augmentation - Replace upside down ](#19)

```python
print(textaugs.replace_upside_down(txt))
```

```python
print(textaugs.split_words(txt))
```
